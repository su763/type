```python
# i. Clean the data by removing sensitive columns [cite: 8]
columns_to_drop = ['school', 'sex', 'age', 'address']
df_cleaned = df.drop(columns=columns_to_drop)

print("\nColumns after dropping sensitive ones:")
print(df_cleaned.columns)

# ii. Create dummy variables for all remaining categorical attributes [cite: 9]
df_dummified = pd.get_dummies(df_cleaned, drop_first=True)

print("\nColumns after creating dummy variables:")
print(df_dummified.columns)
print("\nFirst 5 rows of the fully prepared data:")
print(df_dummified.head())
```

```python
# i. Define the target (Y) and predictor (X) variables [cite: 11]
Y = df_dummified['G3']  # Target variable (final grade)
X = df_dummified[['G2']] # Predictor variable (second period grade)

# Add a constant (intercept) to the predictor variable
# This is required by the statsmodels OLS function
X = sm.add_constant(X)

# ii. Build the Ordinary Least Squares (OLS) Linear Regression model [cite: 12]
model = sm.OLS(Y, X).fit()
```

```python
# i. Display the regression summary [cite: 14]
print(model.summary())
```

```python
# Get predictions and residuals from the model
df_dummified['predicted_G3'] = model.predict(X)
df_dummified['residuals'] = model.resid

# i. Perform the Breusch-Pagan test for heteroscedasticity [cite: 17]
# Null hypothesis (H0): The residuals are homoscedastic (constant variance).
# Alternative hypothesis (H1): The residuals are heteroscedastic (non-constant variance).
bp_test = het_breuschpagan(model.resid, model.model.exog)
labels = ['Lagrange Multiplier statistic', 'p-value', 'f-value', 'f p-value']
bp_results = dict(zip(labels, bp_test))

print("\nBreusch-Pagan Test Results:")
print(bp_results)

# Interpretation of the Breusch-Pagan test
if bp_results['p-value'] < 0.05:
    print("\nResult: The p-value is less than 0.05, so we reject the null hypothesis.")
    print("This suggests the presence of heteroscedasticity.")
else:
    print("\nResult: The p-value is greater than 0.05, so we fail to reject the null hypothesis.")
    print("This suggests the residuals are homoscedastic.")


# ii. Calculate Variance Inflation Factor (VIF) [cite: 18]
# Note: VIF is used to detect multicollinearity among predictors.
# Since we only have one predictor ('G2'), VIF is not applicable here.
# A VIF score can only be calculated with two or more predictor variables.
print("\nVariance Inflation Factor (VIF):")
print("VIF is not applicable for a simple linear regression model with only one predictor.")


# iii. Plot residuals to check assumptions [cite: 19]
plt.figure(figsize=(10, 6))
sns.residplot(x=df_dummified['predicted_G3'], y=df_dummified['residuals'], lowess=True, line_kws={'color': 'red', 'lw': 2})
plt.title('Residuals vs. Fitted Values Plot')
plt.xlabel('Fitted Values (Predicted G3)')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()

# Interpretation of the residual plot:
# We are looking for a random scatter of points around the horizontal line at y=0.
# If there is a clear pattern (like a curve or a cone shape), it indicates a problem
# with the model's assumptions (e.g., non-linearity or heteroscedasticity).
# In this case, the points appear relatively random, which is good.
```
